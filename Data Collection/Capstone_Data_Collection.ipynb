```python
# Image Samples

import matplotlib.pyplot as plt
from PIL import Image

file_path = '/content/drive/My Drive/Capstone Data Collection/extracted_sample_images/'


jpeg_files = [f for f in os.listdir(file_path) if f.endswith('.jpg')]

jpeg_files = jpeg_files[:10]

fig, axes = plt.subplots(2, 5, figsize=(15, 6))
axes = axes.flatten()

for ax, file in zip(axes, jpeg_files):
    img_path = os.path.join(file_path, file)

    try:
        img = Image.open(img_path)
        img = img.convert('RGB')
        ax.imshow(img)
    except Exception as e:
        print(f"Error loading image {file}: {e}")
        ax.text(0.5, 0.5, 'Image Error', ha='center', va='center')

    ax.axis('off')

plt.tight_layout()
plt.show()
```
