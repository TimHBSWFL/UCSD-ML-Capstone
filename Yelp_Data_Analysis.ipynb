```python
import json
import pandas as pd
from google.colab import files
```

```python
uploaded = files.upload()
```

```python
filename = next(iter(uploaded.keys()))
print(filename)
```

```python
data = []

with open(filename, 'r') as file:
    try:
      data = json.load(file)
    except json.JSONDecodeError:
      file.seek(0)
      for line in file:
        data.append(json.loads(line))

```

```python
df = pd.DataFrame(data)
df.head()
```
